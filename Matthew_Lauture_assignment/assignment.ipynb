1.
Wickham wants us to think more deeply about data cleaning, not just removing missing values. He asks what the data should look like to know it’s clean. He suggests that each row is an observation, each column is a variable, and different types of data should be in separate tables. This helps us understand how to structure and clean our data effectively.

2.
Data cleaning takes up a lot of time and effort in data analysis, but it isn't studied much as a skill. The "tidy data standard" aims to set clear guidelines for how to clean data, making the process easier. The idea is that if everyone follows the same steps and knows the goal, it becomes simpler to clean and organize the data properly.

3. Dysfunctional, messy data creates unique challenges, while clean data tends to be more straightforward. Often, data is structured for the creator’s needs, making it hard to manipulate in R.

The idea of a "data frame/matrix" seems simple—rows as observations and columns as variables—but it's a crucial choice that should be made thoughtfully. For example, in county-year data, each observation is a county-year combination. If someone misinterprets this as just a county, it can lead to poor decisions in cleaning and organizing the data.

4.
A dataset is a group of values, which can be numbers or categories. Each value belongs to both a variable and an observation. A variable includes values that measure the same attribute (like height or color), while an observation is a collection of values that describe a single instance (like a specific person's height and color).

5. In tidy data, each variable is a column, each observation is a row, and different types of data are in separate tables. If data doesn't follow this structure, it's considered messy.

6. Column headers are values: In a dataset showing unemployment by county over time, columns represent years (like 2012), which are values, not variable names like "Year."

Multiple variables in one column: Dates like month-day-year combine three pieces of information (month, day, year) and often need to be changed to a different format for analysis.

Variables in both rows and columns: Sometimes, the same type of information appears in both rows and columns, which can be confusing.

Different types of units in one table: Data about different groups (like entrepreneurs and workers) are mixed in one file instead of having separate datasets.

Single units in multiple tables: The same data might be repeated across tables, making it messy and needing cleaning.

4. Missing values, not at random: For the pretrial data covered in the lecture, clean the ImposedSentenceAllChargeInContactEvent variable as well as you can, and explain the choices you make. (Hint: Look at the SentenceTypeAllChargesAtConvictionInContactEvent variable.)


```python
length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']

#
length = pd.to_numeric(length,errors='coerce') # Coerce to numeric
length_NA = length.isnull() # Create a missing dummy
print( np.sum(length_NA),'\n') # a lot of missing values out of total dataset

print( pd.crosstab(length_NA, type), '\n') # Category 4 is cases where the charges were dismissed

length = length.mask( type == 4, 0) # Replace length with 0 when type ==4
length = length.mask( type == 9, np.nan)
length_NA = length.isnull()
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n') # 274 missing, much better

df['ImposedSentenceAllChargeInContactEvent'] = length
del length, type # Delete length/variables
```
